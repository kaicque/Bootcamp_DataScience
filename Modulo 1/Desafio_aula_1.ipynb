```python
import pandas as pd
dados = pd.read_csv("/content/A213527189_28_143_208.csv", encoding="iso-8859-1",
                    skiprows = 3, sep=";", skipfooter = 12,
                    thousands=".", decimal=",")
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

#Desafio 01: escolher um título mais descritivo, que passe a mensagem adequada.

```python
ax = dados.plot(x="Unidade da Federação",y="2008/Ago", kind="bar", figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R${x:,.2f}"))
plt.title("Despesas de procedimentos hospitalares por local de internação contabilizado por saida de caixa")
plt.show()
```

#Desafio 02: faça a mesma análise pro mês mais recente

```python
ax = dados.plot(x="Unidade da Federação",y="2021/Mar", kind="bar", figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R${x:,.2f}"))
plt.title("Despesas de procedimentos hospitalares por local de internação contabilizado por saida de caixa")
plt.show()
```
