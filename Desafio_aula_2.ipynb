```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
dados = pd.read_csv("/content/A001259189_28_143_208.csv", encoding="iso-8859-1",
                    skiprows = 3, sep=";", skipfooter = 12,
                    thousands=".", decimal=",")
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0,"Unidade da Federação")
usaveis = dados[colunas_usaveis]
usaveis = usaveis.set_index("Unidade da Federação")
usaveis = usaveis.drop("Total", axis=1)
```

#Desafio 01: Reposicionar a legenda do gráfico em uma posição mais adequada

```python
ax = usaveis.T.plot(figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R${x:,.2f}"))
plt.title("Despesas de procedimentos hospitalares por local de internação contabilizado por data da internação")
plt.legend(loc=0, bbox_to_anchor=(1, 1))
plt.show()
```

#Desafio 02: Plotar o gráfico de linha com apenas 5 estados de sua preferência

```python
selecionados = usaveis.loc[["50 Mato Grosso do Sul","13 Amazonas","41 Paraná","42 Santa Catarina","43 Rio Grande do Sul"]]
ax = selecionados.T.plot(figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R${x:,.2f}"))
plt.title("Despesas de procedimentos hospitalares por local de internação contabilizado por data da internação")
plt.legend(title= "Estados", loc=0, bbox_to_anchor=(1.05, 1))
plt.show()
```
